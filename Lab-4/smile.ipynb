```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((500,500,3), np.uint8)
```

```python
cv.circle(img,(250,250), 230, (255, 222, 0), -1)

plt.imshow(img)
plt.show()
```

```python
cv.ellipse(img,(150,200),(100,50),90,0,360,000,-1)
cv.ellipse(img,(350,200),(100,50),90,0,360,000,-1)

cv.ellipse(img,(250,340),(100,100),0,5,175,000,8)

plt.imshow(img)
plt.show()
```
