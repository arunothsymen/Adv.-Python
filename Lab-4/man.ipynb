```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((512,700,3), np.uint8)
```

```python
cv.circle(img,(250,150), 60, (0,0,255), 4)
cv.circle(img,(275,130), 5, (0,0,255), 4)
cv.circle(img,(230,130), 5, (0,0,255), 4)
cv.line(img,(250,213),(250,350),(0,0,255),5)
cv.line(img,(250,213),(150,300),(0,0,255),5)
cv.line(img,(250,213),(350,300),(0,0,255),5)
cv.line(img,(250,350),(150,450),(0,0,255),5)
cv.line(img,(250,350),(350,450),(0,0,255),5)
cv.ellipse(img,(252,155),(40,40),0,0,180,255, -1)

plt.imshow(img)
plt.show()
```
