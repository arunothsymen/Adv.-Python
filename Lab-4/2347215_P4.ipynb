  **OpenCV library**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**1. Image Opening and Display**

```python
input_image = cv2.imread("Input_image.jpg")
plt.subplot(2, 3, 1), plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)), plt.title("Original Image")
plt.axis('off')
```

**2. Resize to 400*400**

```python
resized_image = cv2.resize(input_image, (400, 400))
plt.subplot(2, 3, 2), plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)), plt.title("Resized Image")
plt.axis('off')
```

**3. Grayscale Conversion**

```python
gray_scale = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
plt.subplot(2, 3, 3), plt.imshow(gray_scale, cmap='gray'), plt.title("Grayscale Image")
plt.axis('off')
```

**4. Crop and Display Grayscale**

```python
roi = gray_scale[100:300, 100:300]
plt.subplot(2, 3, 4), plt.imshow(roi, cmap='gray'), plt.title("Cropped Image")
plt.axis('off')
```

**5. Image Rotation**

```python
rotation_matrix = cv2.getRotationMatrix2D((100, 100), 45, 1)
rotated_image = cv2.warpAffine(roi, rotation_matrix, (200, 200))
plt.subplot(2, 3, 5), plt.imshow(rotated_image, cmap='gray'), plt.title("Rotated Image")
plt.axis('off')
```

**6. Applying Filters**

```python
gauss_blur = cv2.GaussianBlur(rotated_image, (5, 5), 0)
plt.subplot(2, 3, 6), plt.imshow(gauss_blur, cmap='gray'), plt.title("Blurred Image")
plt.axis('off')
```

**7. Saving the Result**

```python
# Create subplots to display all images
fig, axs = plt.subplots(2, 3, figsize=(16, 8))

# Plot each image in a subplot
axs[0, 0].imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title("Original Image")
axs[0, 0].axis('off')


axs[0, 1].imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
axs[0, 1].set_title("Resized Image")
axs[0, 1].axis('off')


axs[0, 2].imshow(gray_scale, cmap='gray')
axs[0, 2].set_title("Grayscale Image")
axs[0, 2].axis('off')

axs[1, 0].imshow(roi, cmap='gray')
axs[1, 0].set_title("ROI Image")
axs[1, 0].axis('off')


axs[1, 1].imshow(rotated_image, cmap='gray')
axs[1, 1].set_title("Rotated Image")
axs[1, 1].axis('off')


axs[1, 2].imshow(gauss_blur, cmap='gray')
axs[1, 2].set_title("Blurred Image")
axs[1, 2].axis('off')

plt.tight_layout()
plt.savefig("Output_image.jpg")
plt.show()
```
